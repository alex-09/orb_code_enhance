```python

def ransac_test(nfeatures=5000, estimator=cv2.RANSAC):# Find matches (4th param could use optimal_nfeature or a fixed value)
    matches, visuals = find_matches(query_image, query_filename, test_imgs, nfeatures, estimator) # this returns the matches and the array of visuals for comparison

    # for v in visuals:
    #     visualize_and_save_matches(v[0], v[1], v[2], v[3], v[4], "RANSAC"+v[5], v[6])

    for match in matches:
        print(match)

    return (matches[0][2], matches[0][4])
```
